```python
import pandas as pd
```

```python
df = pd.read_csv(r'shedulers.csv', encoding='cp1251', sep=';')
df.head(10)
```

```python
working_days = pd.DataFrame(columns=['Name', 'Date_begin', 'Date_end'])
```

```python
from datetime import datetime, timedelta

def add_working_days(start, end, dataset=working_days):
  start = datetime.strptime(start, '%d.%m.%Y')
  end = datetime.strptime(end, '%d.%m.%Y')
  for index, row in df.iterrows():
      name = row[0]
      schedule = row[1]
      date_begin = datetime.strptime(row[2][:10], '%d.%m.%Y')
      date_end = datetime.strptime(row[3][:10], '%d.%m.%Y')
      
      if (date_begin > start):
        start_day = date_begin
      else:
        start_day = start

      if (date_end < end):
        end_day = date_end
      else:
        end_day = end

      if (start_day <= end_day):
      
        interval_days = (end_day - start_day).days
        first_day = (start_day - date_begin).days
        
        index_schedule  = first_day
        n = interval_days
        i = 0
            
        while (i <= n) :

          day_val = schedule[index_schedule %len(schedule)]
        
          if (day_val == "н"):

              work_start = start_day + timedelta(days=i,hours=20)
              work_end = start_day + timedelta(days=i+1,hours=8)
              new_row = {'Name': name, 'Date_begin': work_start, 'Date_end': work_end}
              dataset = dataset.append(new_row, ignore_index=True)
              i += 1

          if (day_val == "д"):

              work_start = start_day + timedelta(days=i,hours=8)
              work_end = start_day + timedelta(days=i,hours=20)
              new_row = {'Name': name, 'Date_begin': work_start, 'Date_end': work_end}
              dataset = dataset.append(new_row, ignore_index=True)

              day_val_next = schedule[(index_schedule+1) %len(schedule)]
              
              if (day_val_next != 'н'):
                  i +=1

          if (day_val == "с"):

              work_start = start_day + timedelta(days=i,hours=8)
              work_end = start_day + timedelta(days=i+1,hours=8)
              new_row = {'Name': name, 'Date_begin': work_start, 'Date_end': work_end}
              dataset = dataset.append(new_row, ignore_index=True)
              i += 1
          
          if (day_val == "в"):
              i += 1;
          index_schedule += 1
  return dataset
```

```python
working_days = add_working_days('01.01.2019','08.01.2019')
```
